```python
import pandas as pd
```

```python
file_path = "../csv/Error 1.csv"
data = pd.read_csv(file_path)
error_code_frequency_index = data[data.iloc[:, 0].str.contains("ERROR CODE FREQUENCY")].index[0]

df_ecl = data.iloc[0:error_code_frequency_index, ]
df_ecf = data.iloc[error_code_frequency_index: , ]
```

```python
df_ecl_formatted = df_ecl.iloc[8:].reset_index(drop=True) # Skip some rows that are not required
df_ecl_formatted = df_ecl_formatted.iloc[:, 0].str.split(';', expand=True) # Split the text into separate with ';' as delimiter 
df_ecl_formatted.columns = df_ecl_formatted.iloc[0] # Reassign columns 
df_ecl_formatted = df_ecl_formatted.drop(df_ecl_formatted.columns[[0, -1]], axis=1) # Drop unrequired columns
df_ecl_formatted = df_ecl_formatted.iloc[1:].reset_index(drop=True) # Reset the index
df_ecl_formatted
```

```python
df_ecf_formatted = df_ecf.iloc[1:].reset_index(drop=True) # Skip some rows that are not required
df_ecf_formatted = df_ecf_formatted.iloc[:, 0].str.split(';', expand=True) # Split the text into separate with ';' as delimiter 
df_ecf_formatted.columns = df_ecf_formatted.iloc[0] # Reassign columns 
df_ecf_formatted = df_ecf_formatted.drop(df_ecf_formatted.columns[-1], axis=1) # Drop unrequired columns
df_ecf_formatted = df_ecf_formatted.iloc[1:].reset_index(drop=True) # Reset the index
df_ecf_formatted
```
